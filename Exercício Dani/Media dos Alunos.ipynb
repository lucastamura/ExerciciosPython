# Faça um programa que leia informações sobre as notas das provas e dos trabalhos realizados pelos alunos de uma determinada disciplina e informe:
* A média final de cada aluno
* O status do aluno: Aprovado, Recuperação ou Reprovado
* Quantos alunos passaram
* Quantos alunos em recuperação
* Quantos alunos reprovados
* Média da classe


Para exibir as informações acima, considerar as seguintesregras:
* Serão realizadas 2 provas: P1, P2
* A prova substitutiva PS é opcional, e substituiu o valor da menor nota entre P1, P2
* A média final das provas, MP, é a média aritmética entre as provas
* A média final dos trabalhos, MT é a média aritmética entre os trabalhos
* A média final, MF, na disciplina corresponde a 60% a média final das provas e 40% a média final dos trabalhos

### Em função da média final, os alunos são classificados em:
* Aprovado: MF >= 6,0
* Recuperação: 4 =< MF<6
* Reprovado MF < 4
Formato de entrada:

Dever-se ler quantos alunos fazem parte da disciplina e quantos trabalhos foram passados. Quando o aluno não realizou a prova substitutiva, deve-se  nformar um valor negativo. Verificar o exemplo de utilização ao final deste documento.


```python
# 1º - Saber a quantidade de alunos e de trabalhos

qtd_alunos = int(input("Informe a quantidade de alunos:"))
qtd_trabalhos = int(input("Informe a quantidade de trabalhos:"))


# 2º Definir as variáveis a serem apresetnadas no geral
qtd_aprovados = 0
qtd_recuperacao = 0
qtd_reprovados = 0
media_classe = 0


# 3º - Criar um while para calcular a média final de cada aluno
i = 1 # Contador do while

while (i <= qtd_alunos): #Ira repedir as perguntas na mesma quantidade de alunos
  
  #Definir variáveis a serem calculadas na média das provas
  media = 0

  #Definir variáveis a serem calculadas na média dos trabalhos
  media_trabalhos = 0
  media_final = 0
  nota_trabalho = 0
  soma_trabalhos = 0
  


  # Calcular a média das notas das provas

  print("============ Notas do aluno", i,"=================")

  #Entrada de dados das notas das provas
  P1 = float(input("Informe a nota da P1:"))
  P2 = float(input("Informe a nota da P2:"))
  PS = float(input("Informe a nota da PS:"))

  #Saída das notas das provas
  print("Nota das provas:")
  print("P1 = ", P1)
  print("P2 = ", P2)
  print("PS = ", PS)

  #Inicio do processamento de dados das notas das provas
  # Se a pessoa não fez PS ela ira colocar uma valor negativo na nota da PS
  #Caso a pessoa não fez a PS calular a média das P1 + P2
  if(PS < 0):
    media = (P1 + P2)/2

  #Se a pessoa fez a PS precisamos validar se a PS ira substituir a P1 ou a P2  
  elif(PS >= 0):

    #PS irá substituir a P1 caso PS >= P1
    if (PS >= P1):
      media = (PS + P2) / 2

    #PS irá substituir a P2 caso PS >= P2
    elif (PS >= P2):
      media = (P1 + PS) / 2
  #Termino do processamento de dados das notas das provas


  #Início do processamento de dados dos trabalhos
  j = 1 # Contador do while

  while (j <= qtd_trabalhos): #ira perguntar as notas das quantidade de trabalhos inseridas inicialmente

    #Entrada de dados dos trabalhos
    trabalho = float(input("Informe a nota do trabalho: "))

    #Saida de dados das notas dos trabalhos
    print("Nota do Trabalho", j ," = ", trabalho)

    #Somando todas as notas para calulcar a média
    soma_trabalhos = soma_trabalhos + trabalho

    #Incremento do contador
    j = j + 1
  #Termino do processamento de dados dos trabalhos

  
  #Calcular a média das notas dos trabalhos
  media_trabalhos = soma_trabalhos / qtd_trabalhos

  #Calcular a média final (Media dos trabalhos = 40% da nota e Media das provas = 60% da nota)
  media_final = media_trabalhos*0.4 + media * 0.6

  #Saída de dados das médias
  print("Média de trabalhos do aluno = ", media_trabalhos)
  print("Média do aluno ",i, " = ",media)
  print("Média Final = ", media_final)

  #Processamento de dados da aprovação do aluno

  #Aluno aprovado:
  if (media_final >= 6):
    print("Aluno aprovado")

    #Incrementar o aluno na lista de aprovados
    qtd_aprovados = qtd_aprovados + 1

    #Somando a média do aluno com a média da classe
    media_classe = media_classe + media_final

  #Aluno em recuperação:
  elif (media_final >=4 and media_final <= 6 ):
    print("Aluno em recuperação")

    #Incrementar o aluno na lista de recuperação
    qtd_recuperacao = qtd_recuperacao + 1

    #Somando a média do aluno com a média da classe
    media_classe = media_classe + media_final

  #Aluno reprovado
  elif (media_final < 4 ):
    print("Aluno reporovado")

    #Incrementar o aluno na lista de reprovados
    qtd_reprovados = qtd_reprovados + 1

    #Somando a média do aluno com a média da classe
    media_classe = media_classe + media_final

  #incremento do contador
  i = i + 1
#Termino do processamento de dados das notas


#Calculando a média da classe
media_alunos = media_classe / qtd_alunos


#Saida dos dados gerais da sala
print("Alunos aprovados: ", qtd_aprovados )
print("Alunos em recuperação: ", qtd_recuperacao)
print("Alunos reporovados: ", qtd_reprovados)
print("Media da classe: ", media_alunos)
```
